```python
import numpy as np
import pandas as pd

df = pd.read_csv("A3.csv")
print("-- DATA PREVIEW --")
print(df.head())
print("\n-- INFO --")
print(df.info())
```

```python
# Summary statistics
print("\n-- DESCRIPTIVE STATISTICS --")
print(df.describe())

# Median, Mode
print("\n-- MEDIAN --")
print(df.select_dtypes(include="number").median)

print("\n-- MODE --")
print(df.mode(numeric_only=True).iloc[0])
```

```python
import matplotlib.pyplot as plt

# Histograms
df.hist(figsize=(10, 6))
plt.suptitle("Histograms of Numerical Columns")
plt.show()

# Boxplots
plt.figure(figsize=(10, 6))
plt.boxplot(df.select_dtypes(include="number"),vert=False)
plt.title("Boxplots of Numerical Columns")
plt.show()
```
