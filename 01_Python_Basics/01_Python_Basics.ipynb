```python
#1.PRIME NUMBER

def isprime(num): 
    if n<2 :
        return False
    for i in range (2,int(n**0.5)+1):
        if(n%i==0):
            return False
    return True

n=int(input("number: "))
if(isprime(n)) :
    print("prime")
else :
    print("not prime")
    
```

```python
#2.PRODUCT OF RANDOM NUMBERS

import random 
n1=random.randint(1,10)
n2=random.randint(1,10)
ca=n1*n2
ans=int(input(f"Enter product of {n1} X {n2} : "))
if ca==ans :
    print("WELLDONE :)")
else :
    print(f"Oops :( The correct answer is :{ca}") 
```

```python
#3.SQUARES OF EVEN OR ODD NUMBERS

str=input("CHOOSE EVEN OR ODD : ").lower()
for i in range(100,201):
    if(str=="even" and i%2==0):
        print(f"{i}^2 = {i**2}")
    elif(str=="odd" and i%2!=0):
        print(f"{i}^2 = {i**2}")
        
```

```python
#4.HOSPITAL BILLING SYSTEM

days=int(input("No. of days: "))
test=int(input("No. of test: "))
medi=int(input("medicine charge: "))
total=1500+(2000*days)+(300*test)+medi
if total >10000:
    fbill=total-(total*0.10)
    print(f"YOUR FINAL BILL : {fbill}")
else:
    print(f"YOUR FINAL BILL : {total}")
    
```
