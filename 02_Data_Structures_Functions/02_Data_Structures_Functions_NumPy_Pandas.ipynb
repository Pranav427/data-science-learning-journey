```python
#1.LIST OPERATIONS

list = [10, 20, 30, 40, 50]
print(f"ORIGINAL LIST elements : {list}")

print("-- APPEND --")
list.append(60)
print(list) 

print("-- EXTEND --")
list.extend([70, 80])
print(list)

print("-- INSERT --")
list.insert(0, 100)
print(list) 

print("-- REMOVE --")
list.remove(70)
print(list) 

print("-- POP --")
list.pop()
print(list)  

print("-- INDEX --")
print(list.index(40))

print("-- COUNT --")
print(list.count(10)) 

print("-- SORT --")
list.sort()
print(list)

print("-- REVERSE --")
list.reverse()
print(list) 

print("-- CLEAR --")
list.clear()
print(list) 
```

```python
#2.TUPLE IMMUTABILITY

stds=("IYLU","IYBUN","BUNNY")
print(stds)
print(len(stds))
try:
    stds[1]="PRANAV" #immutable
    print(stds)
except TypeError as e:
    print("Error: ",e)
```

```python
# 11.EXPLORING PANDAS


data = {
    'Name': ['iylu','pranav','bunny','praneetha','dhanush','abhishek','prem','tharun','sudheer','kiran'],
    'Age': [21, 22, 20, 21, 22, 21, 23, 23, 24, 23],
    'City': ['HYD','CHE','MUM','BLR','PUN','HYD','CHE','HYD','DEL','MUM'],
    'Salary': [25000,30000,28000,35000,30000,np.nan,30000,25000,np.nan,28000]
}
df=pd.DataFrame(data)
print("-- DATA -- \n",df)
print("-- INFO -- \n",df.info())
print("-- DESCRIBE -- \n",df.describe())
print("-- NAME and CITYc--",df[["Name","City"]])
udf=df.drop("City",axis=1)
print("\n-- AFTER DROPPING CITY --\n", udf)
udf['Salary']=udf['Salary'].fillna(udf["Salary"].mean())
print("-- MISSING VALUES FILLED WITH MEAN -- \n",udf)
uqdf=udf.drop_duplicates()
print("-- UNIQUE -- \n",uqdf)
      

```
