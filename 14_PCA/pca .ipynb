```python
# Remove target column (Type)
x = df.drop('Type', axis=1)
x
```

```python
# Standardization

from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x = ss.fit_transform(x)

# PCA (Dimensionality Reduction)
from sklearn.decomposition import PCA
pca = PCA()
x_pca = pca.fit_transform(x)
```

```python
# Explained variance plot
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.show()
```

```python
"""
PCA simplifies high-dimensional data.
It reduces noise and improves model efficiency.
Clustering helps group similar data points.
K-Means works well for clearly separated clusters.
DBSCAN is useful for detecting outliers and irregular clusters.
PCA combined with clustering improves overall analysis quality.

"""
```

```python
# Use first 2 principal components
pca = PCA(n_components=2)
x_pca_2 = pca.fit_transform(x)
```

```python
# K-Means on Original Data
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

scores = []

for i in range(2,11):
    km = KMeans(n_clusters=i)
    labels_original = km.fit_predict(x)
    scores.append(silhouette_score(x, labels_original))
    
# Silhouette score plot
import matplotlib.pyplot as plt
plt.scatter(x=range(2,11),y=scores)
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.title("K-Means Silhouette Score (Original Data)")
plt.plot(range(2,11),scores,color='red')
plt.show()
```

```python
# Final K-Means with k = 3
km = KMeans(n_clusters=3)
labels_original = km.fit_predict(x)
print("Silhouette Score (Original Data):", silhouette_score(x, labels_original))

# K-Means on PCA Data
kmeans_pca = KMeans(n_clusters=3)
labels_pca = kmeans_pca.fit_predict(x)
print("Silhouette Score (PCA Data):",silhouette_score(x_pca_2, labels_pca))
```

```python
# Visualization
plt.scatter(x_pca_2[:, 0], x_pca_2[:, 1], c=labels_pca)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("K-Means Clustering on PCA Data")
plt.show()
```

```python
# DBSCAN on PCA Data
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score

# DBSCAN does not need number of clusters
dbscan = DBSCAN(eps=0.6, min_samples=5)
db_labels = dbscan.fit_predict(x_pca_2)

# Count noise points
noise_points = np.sum(db_labels == -1)
print("Number of noise points:", noise_points)

# Visualization
plt.scatter(x_pca_2[:, 0], x_pca_2[:, 1], c=db_labels)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("DBSCAN Clustering on PCA Data")
plt.show()

```

```python
# Silhouette Score for DBSCAN
unique_labels = set(db_labels)
if len(unique_labels) > 1 and -1 not in unique_labels:
    print("Silhouette Score (DBSCAN):",
          silhouette_score(x_pca_2, db_labels))
else:
    print("Silhouette score not reliable for DBSCAN due to noise points.")
```

```python
# K-Means on original data gave a low silhouette score ~0.28594199657074876.
# After PCA, the silhouette score improved to ~0.56.
# PCA reduced noise and improved cluster separation.
# DBSCAN identified natural clusters and detected noise.
```
