```python
#1.ESTIMATION AND CONFIDENCE INTERVALS
```

```python
#Importing required libraries
import numpy as np
import scipy.stats as stats
import math
```

```python
# Given data in millions
data = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48,1.20, 1.33, 1.18, 1.22, 1.29])

# Basic statistics
n = len(data)
mean = np.mean(data)
std = np.std(data, ddof=1)
alpha = 0.01   # For 99% confidence
```

```python
#Part (a): Confidence Interval using Sample Standard Deviation
t_crit = stats.t.ppf(1 - alpha/2, n - 1)
t_margin = t_crit * (std / math.sqrt(n))
ci_t = (mean - t_margin, mean + t_margin)

print("99% CI (t-distribution):", np.round(ci_t, 3))
```

```python
#Part (b): Confidence Interval using Known Population Std Dev
pop_std = 0.2
z_crit = stats.norm.ppf(1 - alpha/2)
z_margin = z_crit * (pop_std / math.sqrt(n))
ci_z = (mean - z_margin, mean + z_margin)

print("99% CI (z-distribution, sigma=0.2):", np.round(ci_z, 3))
```

```python
# 2.HYPOTHESIS TESTING (Z-TEST)
```

```python
# Given Information
mean_X = 600    #mu               
std_dev_X = 25  #sigma              
sample_mean_cost = 3050        
sample_size = 25               
significance_level = 0.05    

theoretical_mean_cost = 1000 + 5 * mean_X
print('Theoretical population mean : Rs.',theoretical_mean_cost)
```

```python
#Defining Hypotheses
# Null Hypothesis (H0): μ = 4000 → No increase in mean cost
# Alternative Hypothesis (H1): μ > 4000 → Mean cost has increased
```

```python
#Standard Error 
standard_error_cost = 5 * (std_dev_X / np.sqrt(sample_size))
#Compute Z Statistic
z_statistic = (sample_mean_cost - theoretical_mean_cost) / standard_error_cost
#Critical Value 
z_critical_value = stats.norm.ppf(1 - significance_level)

print(f"\nZ-statistic: {z_statistic:.3f}")
print(f"Z-critical (one-tailed, alpha={significance_level}): {z_critical_value:.3f}")
```

```python
# Step 6: Decision
if z_statistic > z_critical_value:
    print("The average weekly cost has increased — reject the null hypothesis (H0).")
else:
    print("The average weekly cost has not increased — fail to reject the null hypothesis (H0).")
```
